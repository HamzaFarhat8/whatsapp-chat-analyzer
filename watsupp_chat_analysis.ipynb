```python
import re
import pandas as pd
```

```python
f=open("WhatsApp Chat with Chivalrous gentlemen.txt",'r',encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
# pattern = r"\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s?(?:am|pm)?\s-\s"
pattern = r"\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s*(?:am|pm)\s-\s"
```

```python
messages=re.split(pattern, data)
len(messages)
```

```python
dates=re.findall(pattern,data)
dates
```

```python
data = data.replace("\u202f", " ")
```

```python
print(len(messages))
print(len(dates))
```

```python
min_len = min(len(messages), len(dates))

messages = messages[:min_len]
dates = dates[:min_len]
```

```python
df = pd.DataFrame(
    list(zip(messages, dates)),
    columns=['user_message', 'message_date']
)
```

```python
df['message_date'] = df['message_date'].str.replace('\u202f', ' ', regex=False)

df['message_date'] = pd.to_datetime(
    df['message_date'],
    format='%d/%m/%Y, %I:%M %p - '
)

df.rename(columns={'message_date': 'date'}, inplace=True)

print(df.head())
```

```python
print("Dates:", len(dates))
print("Messages:", len(messages))
```

```python
df.head()
```

```python
import re

users = []
messages = []

for message in df['user_message']:
    message = str(message)  # safety against NaN / None

    entry = re.split(r'^([^:]+):\s', message, maxsplit=1)

    if len(entry) == 3:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(message)
```

```python
df.head()
```

```python
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)

df.head()
```

```python
df['year']=df['date'].dt.year
```

```python
df['hour']=df['date'].dt.hour
```

```python
df['minute']=df['date'].dt.minute
```
